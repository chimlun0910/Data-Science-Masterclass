**Q1. Explain GET and POST methods.**

- GET is used to request data from a specified resource. GET request can be cached and remains in browser history. GET is not used when dealing with sensitive data.
- POST is used to send data to a server to create / update a resource. POST cannot be cached and remains in browser history.

**Q2. Why is request used in Flask?**

- The Request in Flask is an object contains all the data sent from a Client to a Server. This data can be recovered using GET / POST Methods.

**Q3. Why is redirect() used in Flask?**

- A redirect() is a method in Flask. When invoked, it returns a response object and redirects the user to another target location with the specified status code.

**Q4. What are templates in Flask? Why is the render_template() function used?**

- Templates are files that contain static data as well as placeholders for dynamic data. A template is rendered with specific data to produce a final document. Flask uses the Jinja template library to render templates.

- This view function uses the render_template() function to render a template file called index.html

```python
from flask import Flask, render_template, request, jsonify

app = Flask(__name__)

@app.route("/")
def home_page():
    return render_template("index.html")

@app.route("/register", methods = ["POST"])
def register():
    if (request.method == "POST"):
        name = request.json["name"].capitalize()
        course_name = request.json["course_name"]
        mail_id = request.json["mail_id"]
        phone_number = request.json["phone_number"]
        result = "Name: {name}, Course: {course_name}, Email: {mail_id}, Phone: {phone_number}".format(name = name, course_name = course_name, mail_id = mail_id, phone_number = phone_number)
        # return render_template("result.html", result = result)
        return jsonify(result)

if __name__ == "__main__":
    app.run(host = "0.0.0.0")
```
